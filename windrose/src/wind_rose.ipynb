```python
from windrose import WindroseAxes
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
import windrose
from google.colab import drive
drive.mount('/content/drive')
```

```python
PATH_DRIVE = 'drive/MyDrive/Seminario-3-master'
PATH_DADOS = '/windrose/'
FILE_WINDROSE = 'dara_wr.csv'
```

```python
df = pd.read_csv(PATH_DRIVE + PATH_DADOS + FILE_WINDROSE)
```

```python
df.columns
```

```python
ax = WindroseAxes.from_ax()

ax.bar(df['Direccion'], df['Magnitud'], normed=True, opening=0.8, edgecolor='white', cmap=cm.hot_r)
# ax.bar(df['Direccion'], df['Magnitud'], opening=0.8, edgecolor='k', bins = np.arange(0,8,1.5), nsector = 8, cmap=cm.hot_r) #, cmap= cm.Spectral_r)
ax.set_legend(loc = 'best')
ax.set_xticklabels(['E', 'NE', 'N', 'NW',  'W', 'SW', 'S', 'SE'])
plt.title('Rosa de los vientos')
#plt.savefig('Rosa.png')
```
