```python
import pandas as pd
import numpy as np
import os
import re
from functools import reduce

from google.colab import drive
drive.mount('/content/drive')

%matplotlib inline
```

```python
PATH_DRIVE = 'drive/MyDrive/Seminario-3-master/'
PATH_FILES = 'dados/'
FILENAME = 'Presion_Atmosferica.csv'
```

```python
df = pd.read_csv(PATH_DRIVE + PATH_FILES + FILENAME, 
                 sep = ',', 
                 date_parser = lambda col: pd.to_datetime(col)
                )
```

```python
### Usar para guardar todas las estaciones
# codigo_estacion_list = df["CodigoEstacion"].unique()

### Usar para estaciones especificas
codigo_estacion_list = [26055110]
```

```python
from pathlib import Path

for cod_estacion in codigo_estacion_list:
  print(f"Guardando estación {cod_estacion}")
  
  df_save : pd.DataFrame = df[df["CodigoEstacion"] == cod_estacion].copy()
  
  name2save = f"Presion_Atmosferica_estacion_{cod_estacion}.csv"
  output_file = PATH_DRIVE + PATH_FILES + f"/{cod_estacion}/{name2save}"

  output_file_path = Path(output_file)
  output_file_path.parent.mkdir(parents=True, exist_ok=True)

  df_save.to_csv(output_file_path, header=True, index=False, sep = ';')
```
